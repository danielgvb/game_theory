# GBM

````python
def simulate_gbm(S0, mu, sigma, T, steps):
    """
    Simulate a Geometric Brownian Motion trajectory.

    :param S0: Initial stock price
    :param mu: Drift coefficient
    :param sigma: Volatility coefficient
    :param T: Time horizon
    :param steps: Number of steps in the simulation
    :return: Simulated stock price path
    """
    dt = T / steps  # time increment
    t = np.linspace(0, T, steps)  # time steps
    W = np.random.standard_normal(size=steps)  # Wiener process
    W = np.cumsum(W) * np.sqrt(dt)  # cumulative sum to simulate Brownian motion
    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)  # GBM formula
    return S, t
````

Get the necessary parameters to simulate the GBM

````python
S0 = 100
mu = df['log_returns'].mean()  * np.sqrt(252)# average return for the period of interest
sigma = df['log_returns'].std() * np.sqrt(252) # observed volatility in the period of interest
T = 1 # one calendar year
steps = 365 # days in a year
````

````python
# Simulate GBM
coffee, time_steps = simulate_gbm(S0, mu, sigma, T, steps)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(time_steps, coffee)
plt.title("Geometric Brownian Motion - Coffee Price Simulation")
plt.xlabel("Time (Years)")
plt.ylabel("Coffee Price")
plt.grid(True)
plt.show()
````

# Montecarlo with Payoffs

````python
# Montecarlo simulation
n_iterations = 10000
payoffs = {'up','down'}
alpha = 0.5 # recovery rate after enforcing contract

#define empty lists

price_deltas = []
rhos = []
omegas =[]
Phis = []
Zetas = []
prices_end = []
for i in range(n_iterations):
    # run a GBM on the coffee price
    coffee, time_steps = simulate_gbm(S0, mu, sigma, T, steps)
    price_delta = coffee[-1] - 100
    price_var = (coffee[-1]/100) - 1
    X = np.random.normal()
    Y = np.random.normal()
    rho = np.abs(X * price_delta) # reputational damage
    omega =np.abs( Y * price_delta) # welfare percieved
    prices_end.append(coffee[-1])

    #fill lists
    price_deltas.append(price_delta)
    rhos.append(rho)
    omegas.append(omega)
    

    # scenarios for delta price movement
    if price_delta >= 0: #price delta is positive
        Phi = -price_delta * 0.7 # assume the fine is not the whole default value
        if price_var  > 0.25:
            Zetas.append(1)
        else:
            Zetas.append(0)
        # if omega > rho: # FNC plays  welfare > reputation > play nothing
            # FNC plays N
    # Coffee grower turn
    else:
        Phi = 0
        Zetas.append(0)
        
    Phis.append(Phi)    
````

````python
df_montecarlo = pd.DataFrame({
    'Price_Delta': price_deltas,
    'Rho': rhos,
    'Omega': omegas,
    'Phi': Phis,
    'Zetas':Zetas,
    'Prices':prices_end})

df_montecarlo.head()
````

````python
df_montecarlo.describe()
````

````python
df_montecarlo_up = df_montecarlo[df_montecarlo['Price_Delta'] >0].copy().reset_index(drop = True)
df_montecarlo_down = df_montecarlo[df_montecarlo['Price_Delta'] <=0].copy().reset_index(drop = True)
df_montecarlo_up.head()
````

### Subgame at node 4: Price up, CF plays D

````python
# create an object that has the moves of FNC
plays_ud = []
for i in range(len(df_montecarlo_up)):
    omega = df_montecarlo_up.iloc[i]['Omega']
    rho = df_montecarlo_up.iloc[i]['Rho']
    # compare omega vs rho
    if omega >= rho:
        plays_ud.append('N')
    else:
        plays_ud.append('E')
df_montecarlo_up['D_outcomes'] = plays_ud
df_montecarlo_up.head()
````

### Subgame at node 5: Price up, CF plays F

````python
# create an object that has the moves of FNC
plays_uf = []
for i in range(len(df_montecarlo_up)):
    price_delta = df_montecarlo_up.iloc[i]['Price_Delta']
    omega = df_montecarlo_up.iloc[i]['Omega']
    rho = df_montecarlo_up.iloc[i]['Rho']
    payoff_n = -price_delta + omega
    payoff_e = -price_delta + rho + (alpha*price_delta)
    # compare payoff that include rho, omega and Delta_price
    if payoff_n >= payoff_e:
        plays_uf.append('N')
    else:
        plays_uf.append('E')
df_montecarlo_up['F_outcomes'] = plays_uf
df_montecarlo_up.head()
````

### Subgame at node 6: Price down, CF plays D

````python
# create an object that has the moves of FNC
plays_dd = []
for i in range(len(df_montecarlo_down)):
    omega = df_montecarlo_down.iloc[i]['Omega']
    rho = df_montecarlo_down.iloc[i]['Rho']
    # compare omega vs rho
    if omega >= -rho:
        plays_dd.append('N')
    else:
        plays_dd.append('E')
df_montecarlo_down['D_outcomes'] = plays_dd
df_montecarlo_down.head()
````

### Subgame at node 7: Price down, CF plays F

````python
# create an object that has the moves of FNC
plays_df = []
for i in range(len(df_montecarlo_down)):
    price_delta = df_montecarlo_down.iloc[i]['Price_Delta']
    omega = df_montecarlo_down.iloc[i]['Omega']
    rho = df_montecarlo_down.iloc[i]['Rho']
    payoff_n = -price_delta + omega
    payoff_e = -price_delta + rho + (alpha*price_delta)
    # compare payoff that include rho, omega and Delta_price
    if payoff_n >= payoff_e:
        plays_df.append('N')
    else:
        plays_df.append('E')
df_montecarlo_down['F_outcomes'] = plays_df
df_montecarlo_down.head()
````

### Subgame at node 2: Price up, Backward Induction for CF

````python
df_montecarlo_up.head()
````

````python
# create an object with the plays of coffee grower
plays_u = []
for i in range(len(df_montecarlo_up)):
    fnc_played_for_d = df_montecarlo_up.iloc[i]['D_outcomes']
    fnc_played_for_f = df_montecarlo_up.iloc[i]['F_outcomes']
    Z = df_montecarlo_up.iloc[i]['Zetas']
    price_delta = df_montecarlo_up.iloc[i]['Price_Delta']
    Phi = df_montecarlo_up.iloc[i]['Phi']
    # check the payoff of playing D
    # for when FNC played N
    if fnc_played_for_d == 'N':
        payoff_d = 0
    else:
        payoff_d = 0
    # compare with payoff of playing F
    if fnc_played_for_f == 'N':
        payoff_f = Z * price_delta
    else:
        payoff_f = (Z * price_delta)  - Phi
    if payoff_d >= payoff_f:
        plays_u.append('D')
    else:
        plays_u.append('F')

df_montecarlo_up['CF'] = plays_u
df_montecarlo_up.head()
````

### Solved Game up

````python
outcomes_up = []
for i in range(len(df_montecarlo_up)):
    cf_plays = df_montecarlo_up.iloc[i]['CF']
    if cf_plays == 'F':
        fnc_plays = df_montecarlo_up.iloc[i]['F_outcomes']
    else:
        fnc_plays = df_montecarlo_up.iloc[i]['D_outcomes']
    final_outcome = cf_plays + ' ' + fnc_plays
    outcomes_up.append(final_outcome)
df_montecarlo_up['Solution'] = outcomes_up
df_montecarlo_up.head()
````

````python
df_montecarlo_up.groupby(['Solution'])[['Solution']].count()
````

### Subgame at node 3: Price down, Backward Induction for CF

````python
df_montecarlo_down.head()
````

````python
# create an object with the plays of coffee grower
plays_d = []
for i in range(len(df_montecarlo_down)):
    fnc_played_for_d = df_montecarlo_down.iloc[i]['D_outcomes']
    fnc_played_for_f = df_montecarlo_down.iloc[i]['F_outcomes']
    Z = df_montecarlo_down.iloc[i]['Zetas']
    price_delta = df_montecarlo_down.iloc[i]['Price_Delta']
    Phi = df_montecarlo_down.iloc[i]['Phi']
    # check the payoff of playing D
    # for when FNC played N
    if fnc_played_for_d == 'N':
        payoff_d = 0
    else:
        payoff_d = 0
    # compare with payoff of playing F
    if fnc_played_for_f == 'N':
        payoff_f = Z * price_delta
    else:
        payoff_f = Z * price_delta
    if payoff_d >= payoff_f:
        plays_d.append('D')
    else:
        plays_d.append('F')

df_montecarlo_down['CF'] = plays_d
df_montecarlo_down.head()
````

### Solved Game down

````python
outcomes_down = []
for i in range(len(df_montecarlo_down)):
    cf_plays = df_montecarlo_down.iloc[i]['CF']
    if cf_plays == 'F':
        fnc_plays = df_montecarlo_down.iloc[i]['F_outcomes']
    else:
        fnc_plays = df_montecarlo_down.iloc[i]['D_outcomes']
    final_outcome = cf_plays + ' ' + fnc_plays
    outcomes_down.append(final_outcome)
    
df_montecarlo_down['Solution'] = outcomes_down
df_montecarlo_down.head()
````

````python
df_sol_down = df_montecarlo_down.groupby(['Solution'])[['Solution']].count().rename(columns = {'Solution':'Count_down'}).reset_index(drop = False)
df_sol_up = df_montecarlo_up.groupby(['Solution'])[['Solution']].count().rename(columns = {'Solution':'Count_up'}).reset_index(drop = False)
df_sol_down
````

````python
df_sol_merge = df_sol_up.merge(df_sol_down, how = 'outer', on = 'Solution').fillna(0)
df_sol_merge
````

## Visualize simulations
